```python
import yaml
from pathlib import Path
from typing import Literal, Dict, List, TypedDict, Optional, Self
from enum import Enum, EnumType, StrEnum

class Version(Enum):
    DARK    = "dark"
    LIGHT   = "light"

base=Path("yaml/")

def col(version : Version) -> Path:
    return base / Path(f"{version.value}-colors.yaml")
def sem_tok_col(version: Version) -> Path:
    return base / Path(f"{version.value}-semanticTokenColors.yaml")
def tok_col(version: Version) -> Path:
    return base / Path(f"{version.value}-tokenColors.yaml")
def output_yaml(version: Version) -> Path:
    return base / Path(f"{version.value}.yaml")
def output_json(version: Version) -> Path:
    return base / Path(f"{version.value}.json")
```

```python
def get_yaml(fp : Path):
    with open(fp, 'r') as stream:
        try:
            data = yaml.safe_load(stream)
            # print(data)
            return data
        except yaml.YAMLError as exc:
            raise Exception(exc)
```
